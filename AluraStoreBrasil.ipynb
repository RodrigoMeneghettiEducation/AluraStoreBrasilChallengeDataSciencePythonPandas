### Importação dos dados



```python
#Importação da biblioteca Pandas.
import pandas as pd

#Link github das urls importadas e do notebook baixado.
#https://github.com/alura-es-cursos/challenge1-data-science/tree/main

#Importação dos arquivos de dados.csv para tratamento e análise.
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

#Declaração e definição de variáveis para leitura dos dados importados.
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

#Escopo de análise a serem efetuadas

#1. Análise do faturamento



###Faturamento total de cada um das lojas.

```python
#Calculando o faturamento total de cada loja, somando os valores da coluna "Preço".
faturamento_loja1 = loja1["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

#Exibindo os valores de faturamento total de cada loja.
print(f"O faturamento total da loja 1 é: R$ {faturamento_loja1}")
print(f"O faturamento total da loja 2 é: R$ {faturamento_loja2}")
print(f"O faturamento total da loja 3 é: R$ {faturamento_loja3}")
print(f"O faturamento total da loja 4 é: R$ {faturamento_loja4}")
```

# 2. Vendas por Categoria

###Quantidade de produtos vendidos por categoria em cada loja

```python
#Calculando a quantidade de produtos vendidos por categoria de cada loja.
qtd_produtos_por_categoria_loja1 = loja1['Categoria do Produto'].value_counts()
qtd_produtos_por_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
qtd_produtos_por_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
qtd_produtos_por_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

#Exibindo os valores de quantidade de produtor vendidos por ategoria de cada loja.
print(f"Lista da quantidade de produtor vendidos por categoria da Loja 1 (Linha Preta): \n\n{qtd_produtos_por_categoria_loja1}\n")
print(f"Lista da quantidade de produtor vendidos por categoria da Loja 2 (Linha Vermelha): \n\n{qtd_produtos_por_categoria_loja2}\n")
print(f"Lista da quantidade de produtor vendidos por categoria da Loja 3 (Linha Azul): \n\n{qtd_produtos_por_categoria_loja3}\n")
print(f"Lista da quantidade de produtor vendidos por categoria da Loja 4 (Linha Verde): \n\n{qtd_produtos_por_categoria_loja4}\n")

#Plotagem de gráfico tipo linhas em sobreposição das 4 lojas para comparação dos valores obtidos.
qtd_produtos_por_categoria_loja1.plot(kind='line', legend='Loja1', ylabel="Quantidade de Produtos", figsize=(15, 5), color ='black');
qtd_produtos_por_categoria_loja2.plot(kind='line', legend='Loja2', ylabel="Quantidade de Produtos", figsize=(15, 5), color ='red');
qtd_produtos_por_categoria_loja3.plot(kind='line', legend='Loja3', ylabel="Quantidade de Produtos", figsize=(15, 5), color ='blue');
qtd_produtos_por_categoria_loja4.plot(kind='line', legend='Loja4', ylabel="Quantidade de Produtos", figsize=(15, 5), color ='green');
```

# 3. Média de Avaliação das Lojas


###Satisfação dos clientes de acordo com a média da avaliação de compras de cada loja

```python
#Calculando a média dos valores de "Avaliação de Compras" de cada loja.
avaliacao_loja1 = loja1["Avaliação da compra"].mean()
avaliacao_loja2 = loja2["Avaliação da compra"].mean()
avaliacao_loja3 = loja3["Avaliação da compra"].mean()
avaliacao_loja4 = loja4["Avaliação da compra"].mean()

#Exibindo os valores de média da "Avaliação de Compras" de cada loja.
print(f"A avaliação média da loja 1 é de {avaliacao_loja1.round(3)}, de um máximo de 5.000")
print(f"A avaliação média da loja 2 é de {avaliacao_loja2.round(3)}, de um máximo de 5.000")
print(f"A avaliação média da loja 3 é de {avaliacao_loja3.round(3)}, de um máximo de 5.000")
print(f"A avaliação média da loja 4 é de {avaliacao_loja4.round(3)}, de um máximo de 5.000")
```

# 4. Produtos Mais e Menos Vendidos

```python
#Contagem dos produtos vendidos em cada loja.
qtd_produtos_loja1 = loja1['Produto'].value_counts()
qtd_produtos_loja2 = loja2['Produto'].value_counts()
qtd_produtos_loja3 = loja3['Produto'].value_counts()
qtd_produtos_loja4 = loja4['Produto'].value_counts()

#Os 5 produtos mais vendidos de cada loja.
top_5_loja1 = qtd_produtos_loja1.head()
top_5_loja2 = qtd_produtos_loja2.head()
top_5_loja3 = qtd_produtos_loja3.head()
top_5_loja4 = qtd_produtos_loja4.head()

#Os 5 produtos menos vendidos de cada loja.
bottom_5_loja1 = qtd_produtos_loja1.tail()
bottom_5_loja2 = qtd_produtos_loja2.tail()
bottom_5_loja3 = qtd_produtos_loja3.tail()
bottom_5_loja4 = qtd_produtos_loja4.tail()
```

```python
#Exibição dos Resultados da loja 1
print("\n   Produtos - 5 Mais Vendidos na Loja 1\n")
for i, (produto, qtd) in enumerate(top_5_loja1.head().items(),1):
  print(f"{i}. {produto}: {qtd} unidades")
print("\n   Produtos - 5 Menos Vendidos na Loja 1\n")
for i, (produto, qtd) in enumerate(bottom_5_loja1.tail().iloc[::-1].items(),1):
  print(f"{-i}. {produto}: {qtd} unidades")
```

```python
#Exibição dos Resultados da loja 2
print("\n   Produtos - 5 Mais Vendidos na Loja 2\n")
for i, (produto, qtd) in enumerate(top_5_loja2.head().items(),1):
  print(f"{i}. {produto}: {qtd} unidades")
print("\n   Produtos - 5 Menos Vendidos na Loja 2\n")
for i, (produto, qtd) in enumerate(bottom_5_loja2.tail().iloc[::-1].items(),1):
  print(f"{-i}. {produto}: {qtd} unidades")
```

```python
#Exibição dos Resultados da loja 3
print("\n   Produtos - 5 Mais Vendidos na Loja 3\n")
for i, (produto, qtd) in enumerate(top_5_loja3.head().items(),1):
  print(f"{i}. {produto}: {qtd} unidades")
print("\n   Produtos - 5 Menos Vendidos na Loja 3\n")
for i, (produto, qtd) in enumerate(bottom_5_loja3.tail().iloc[::-1].items(),1):
  print(f"{-i}. {produto}: {qtd} unidades")
```

```python
#Exibição dos Resultados da loja 4
print("\n   Produtos - 5 Mais Vendidos na Loja 4\n")
for i, (produto, qtd) in enumerate(top_5_loja4.head().items(),1):
  print(f"{i}. {produto}: {qtd} unidades")
print("\n   Produtos - 5 Menos Vendidos na Loja 4\n")
for i, (produto, qtd) in enumerate(bottom_5_loja4.tail().iloc[::-1].items(),1):
  print(f"{-i}. {produto}: {qtd} unidades")
```
